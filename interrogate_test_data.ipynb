```python
import pandas as pd

# Load test data
features_test = pd.read_parquet("test.parquet/date_id=0/part-0.parquet")
responders_test = pd.read_parquet("lags.parquet/date_id=0/part-0.parquet")

```

```python
# Basic information about the dataframes
print("Features DataFrame Info:")
features_test.info()
print("\nResponders DataFrame Info:")
responders_test.info()

# Sample of first few rows
print("\nFeatures Test Head:")
print(features_test.head())
print("\nResponders Test Head:")
print(responders_test.head())

# Summary statistics
print("\nFeatures Test Description:")
print(features_test.describe())
print("\nResponders Test Description:")
print(responders_test.describe())

# Check for missing values
print("\nMissing Values in Features Test:")
print(features_test.isnull().sum())
print("\nMissing Values in Responders Test:")
print(responders_test.isnull().sum())

# Column names and types
print("\nFeatures Test Columns and Types:")
for col in features_test.columns:
    print(f"{col}: {features_test[col].dtype}")
print("\nResponders Test Columns and Types:")
for col in responders_test.columns:
    print(f"{col}: {responders_test[col].dtype}")

# Check unique values in categorical columns (if any)
# Adjust the threshold for what's considered categorical
categorical_threshold = 50
for df, name in [(features_test, "Features"), (responders_test, "Responders")]:
    print(f"\nUnique values in {name} categorical columns:")
    for col in df.columns:
        if df[col].nunique() < categorical_threshold:
            print(f"\n{col}:")
            print(df[col].value_counts())

# Check for duplicates
print("\nDuplicate rows in Features Test:", features_test.duplicated().sum())
print("Duplicate rows in Responders Test:", responders_test.duplicated().sum())
```
